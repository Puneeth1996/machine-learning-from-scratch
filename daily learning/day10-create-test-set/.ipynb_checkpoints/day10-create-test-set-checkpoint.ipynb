## 1. Load example data

We use the California Housing dataset to demonstrate split techniques.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split, StratifiedShuffleSplit

housing = fetch_california_housing(as_frame=True)
df = housing.frame.copy()
df.head()
```

### Add an income category for stratified splitting

We create a new `income_cat` column using `median_income`.

```python
df['income_cat'] = pd.cut(
    df['MedInc'],
    bins=[0., 1.5, 3.0, 4.5, 6.0, np.inf],
    labels=[1, 2, 3, 4, 5]
)
df['income_cat'].value_counts().sort_index()
```

## 2. Random train-test split

This split is random and can change every execution unless we fix the seed.

```python
train_set, test_set = train_test_split(df, test_size=0.2)
print('Train size:', len(train_set))
print('Test size:', len(test_set))
train_set['income_cat'].value_counts(normalize=True).sort_index()
```

## 3. Reproducible split with `random_state`

Using `random_state` ensures the same split every time.

```python
train_set_rs, test_set_rs = train_test_split(df, test_size=0.2, random_state=42)
print('Train size:', len(train_set_rs))
print('Test size:', len(test_set_rs))
print('\nIncome category distribution in test set:')
print(test_set_rs['income_cat'].value_counts(normalize=True).sort_index())
```

## 4. Hash-based splitting

Hash-based splitting uses a stable identifier so the same records stay in the same split.

```python
import hashlib

def test_set_check(identifier, test_ratio, hash=hashlib.md5):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio

def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]

housing_with_id = df.reset_index().rename(columns={'index': 'id'})
train_hash, test_hash = split_train_test_by_id(housing_with_id, 0.2, 'id')
print('Hash-based test size:', len(test_hash))
print('Hash-based train size:', len(train_hash))
```

## 5. Stratified sampling

Stratified sampling preserves the `income_cat` distribution in both train and test sets.

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]

def category_proportions(data):
    return data['income_cat'].value_counts(normalize=True).sort_index()

print('Overall proportions:')
print(category_proportions(df))
print('\nStratified test proportions:')
print(category_proportions(strat_test_set))
print('\nRandom test proportions:')
print(category_proportions(test_set_rs))
```

## 6. Summary and comparison

- Random split: simple, but may not preserve important distributions.
- Reproducible split: same result every time using `random_state`.
- Hash-based split: stable for growing datasets and production use.
- Stratified split: keeps the target or important feature distribution similar.
