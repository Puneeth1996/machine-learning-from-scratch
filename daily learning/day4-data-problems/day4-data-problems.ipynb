# Day 4: Data Problems

In machine learning, good data matters more than fancy algorithms early on. If your data is insufficient, biased, noisy, or missing the right features, even the best model will struggle.

## Common Data Problems

- **Insufficient data**: Not enough examples to learn patterns.
- **Nonrepresentative data**: Training data does not match the real-world cases the model will see.
- **Poor-quality data**: Errors, typos, duplicates, missing values, and inconsistent formats.
- **Irrelevant features**: Columns that do not help prediction or add noise.
- **Why data matters more than algorithms early on**: A simple model on strong, clean data often beats a complex model on weak data.

## Inspecting a Small Dataset

Below is a small example dataset and a quick check for common data-quality issues.

```python
import pandas as pd
from io import StringIO

csv_data = '''
id,age,income,education,score
1,25,50000,Bachelors,88
2,30,NaN,Masters,92
3,,70000,Bachelors,85
4,45,85000,,90
5,28,52000,Bachelors,not available
6,1000,60000,Highschool,70
7,35,55000,Bachelors,88
8,29,54000,Bachelors,88
9,31,53000,Bachelors,88
10,27,52000,Bachelors,88
'''

data = pd.read_csv(StringIO(csv_data))

print('Dataset shape:', data.shape)
print('Preview:')
print(data)

print('Missing values per column:')
print(data.isna().sum())

print('Duplicate rows:')
print(data.duplicated().sum())

print('Data types:')
print(data.dtypes)

print('Potential issues:')
issues = []
if data.shape[0] < 20:
    issues.append('- Insufficient data: only {} rows'.format(data.shape[0]))
if data.isna().any().any():
    issues.append('- Missing values found')
if data.duplicated().any():
    issues.append('- Duplicate rows present')
if data['score'].dtype == object or data['score'].str.contains('[^0-9]', regex=True).any():
    issues.append('- Non-numeric values in score column')
if (data['age'] > 120).any() or (data['age'] < 0).any():
    issues.append('- Implausible age values detected')

for item in issues:
    print(item)
```

## Possible Data-Quality Issues in This Dataset

- Missing age, income, or education values.
- A non-numeric value in the score column.
- An implausible age value of 1000.
- Too few rows for strong generalization.
- Potential duplicate or repeated rows.
- The dataset may not represent all relevant groups or behaviors.
