# Day 6: Testing and Validation

This lesson introduces how to evaluate machine learning models properly. A model that looks good on training data may still fail on new data, so we use validation and testing strategies to measure real performance.

## Why Testing and Validation Matter

- Training error alone is not enough.
- A model may overfit the training set and perform poorly on unseen data.
- Validation helps us choose the best model and hyperparameters.
- The test set gives a final, unbiased estimate of performance.

## Train/Test Split

A simple approach is to split the dataset into two parts:
- Training set: used to fit the model.
- Test set: used to evaluate the final model.

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Create a simple binary classification dataset
X, y = make_classification(n_samples=200, n_features=2, n_informative=2, n_redundant=0, random_state=42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
predictions = model.predict(X_test)

print('Training samples:', len(X_train))
print('Test samples:', len(X_test))
print('Accuracy:', round(accuracy_score(y_test, predictions), 3))
```

## Validation Set and Hyperparameter Tuning

A validation set is used when we want to compare models or tune hyperparameters without using the test set too early.

## Cross-Validation Idea

Cross-validation splits the training data into several folds and repeats training/evaluation across them. This gives a more reliable estimate of model performance.

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification

X2, y2 = make_classification(n_samples=120, n_features=2, n_informative=2, n_redundant=0, random_state=7)

model = DecisionTreeClassifier(random_state=42)
scores = cross_val_score(model, X2, y2, cv=5)

print('Cross-validation scores:', scores)
print('Average score:', round(scores.mean(), 3))
```

## Data Mismatch

Data mismatch happens when the training data does not represent the real-world data the model will face later. This can lead to poor generalization.

## Summary

- Use a train/test split for final evaluation.
- Use a validation set to tune hyperparameters.
- Use cross-validation for a more stable performance estimate.
- Watch for data mismatch to avoid misleading results.
