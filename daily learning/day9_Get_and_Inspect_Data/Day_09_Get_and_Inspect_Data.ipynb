## Why inspect data?
Before training any ML model, understand your data. Good data understanding leads to better models.

```python
import pandas as pd
import matplotlib.pyplot as plt

print("Libraries imported.")
```

## Load Dataset
Download `housing.csv` from the California Housing dataset on Kaggle and place it in `datasets/`.

```python
housing = pd.read_csv("california_housing.csv")
housing.head()
```

### Explanation
`head()` displays the first five rows so you can quickly inspect the dataset.

```python
housing.shape
```

`shape` returns `(rows, columns)`.

```python
housing.info()
```

`info()` shows data types, missing values, and memory usage.

```python
housing.describe()
```

`describe()` summarizes numerical columns.

```python
housing["HouseAge"].value_counts()
```

`value_counts()` shows category frequencies.

```python
housing.isnull().sum()
```

Check for missing values before preprocessing.

```python
housing.hist(figsize=(12,8))
plt.tight_layout()
plt.show()
```

### Histogram Discussion
Look for:
- Skewed distributions
- Outliers
- Wide ranges
- Possible transformations
